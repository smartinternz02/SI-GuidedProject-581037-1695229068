# Data Visualization(MatPlotlib & Seaborn)

# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
days = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
steps = [6745,8765,11456,4567,8799,3456,11567]

plt.plot(days,steps)
```

```python
plt.plot(days,steps,'o-')
```

```python
plt.plot(days,steps,'o-g')
```

```python
# Title the graph
plt.title('Steps count - Last week Vs This week')

# Graph data
days = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
steps = [6745,8765,7456,4567,8799,3456,11567]
lws = [8745,6784,11457,5964,7845,7945,13457]
plt.plot(days,steps,'o-g',lws,'o-m')

# Label the axes
plt.xlabel('Days of the week')
plt.ylabel('Steps Walked')

# To display the grid lines
plt.grid(True)

# Create Legends
plt.legend(['Current week','Last Week'])
```

```python
# Adjust figure size
plt.figure(figsize=(10,5))

# Graph Data
age = [34,36,41,43,28]
name = ['VK','SKY', 'MSD', 'ST', 'HP']
plt.bar(name,age)
```

```python
# Stacked Bar Chart

name = ['Shivam', 'Suryansh', 'Shrikant', 'Vivek']
ot = [29,48,32,46] #y1
ct = [22,34,15,28] #y2

plt.bar(name,ot, label='Open Ticket', width=0.5, color='k')
plt.bar(name,ct, label='Closed Ticket', width=0.5, color='g')

plt.xlabel('Name')
plt.ylabel('Tickets')

plt.legend(['Open Ticket','Closed Ticket'])
```

```python
age = [34,36,41,43,28]
name = ['VK','SKY', 'MSD', 'ST', 'HP']
colors = ['g','r', 'b', 'orange', 'm']
plt.pie(age, [0,0,0.1,0,0], labels = name, autopct="%1.1f%%", shadow=True, colors=colors)
plt.show()
```
